Quantitative Approach to Conceptual Metaphors Based on Fedor Dostoevsky's novel "White Nights" (Belye Nochi). First and Second Chapters

```python
metaphors_general_first_chapter = 91 # Общее количество метафор в первой главе
dead_metaphors_first_chapter = 35 # Количество стертых метафор в первой главе
dormant_metaphors_first_chapter = 33 # Количество обычных, конвенциональных метафор в первой главе
live_metaphors_first_chapter = 23 # Количество новых метафор в первой главе
words_first_chapter = 3484 # Количество слов в первой главе

everyday_metaphors_first_chapter = 47
nature_metaphors_first_chapter = 12
technic_metaphors_first_chapter = 0
violence_metaphors_first_chapter = 0
sport_metaphors_first_chapter = 0
medical_metaphors_first_chapter = 32

metaphors_general_second_chapter = 118 # Общее количество метафор во второй главе
dead_metaphors_second_chapter = 28 # Количество стертых метафор во второй главе
dormant_metaphors_second_chapter = 36  # Количество обычных, конвенциональных метафор во второй главе
live_metaphors_second_chapter = 54 # Количество новых метафор во второй главе
words_second_chapter = 7503 # Количество слов во второй главе

everyday_metaphors_second_chapter = 75
nature_metaphors_second_chapter = 10
technic_metaphors_second_chapter = 0
violence_metaphors_second_chapter = 13
sport_metaphors_second_chapter = 0
medical_metaphors_second_chapter = 20
```

Relative Metaphor Frequency

```python
def RMF(metaphors, words):
    final = metaphors / words
    return final
```

Index of Metaphor Creativity

```python
def IMC(live, dormant, dead, metaphors):
    final = (1*dead + 1.5*dormant + 3*live) / metaphors
    return final
```

Metaphor Frequent Index

```python
def MFI(metaphors, words):
    mean = metaphors / 100
    final = (mean*100) / words
    return final
```

Metaphor Intensity Index

```python
def MII(live, dormant, dead, metaphors):
    final = (live*3 + dormant*2 + dead) / metaphors
    return final
```

Metaphor Content Index

```python
def MCI(everyday, nature, technic, violence, sport, medical, metaphors):
    final = (1*everyday + 2*nature + 3*technic + 4*violence + 5*sport + 6*medical) / metaphors
    return final 
```

Metaphor Power Index Method

```python
def MPI(MFI, MII, MCI):
    final = MFI*MII*MCI
    return final
```

Descriptive Statistics. First Chapter

```python
import pandas as pd
RMF_first = RMF(metaphors_general_first_chapter, words_first_chapter)
IMC_first = IMC(live_metaphors_first_chapter, dormant_metaphors_first_chapter, dead_metaphors_first_chapter, metaphors_general_first_chapter)
MFI_first = MFI(metaphors_general_first_chapter, words_first_chapter)
MII_first = MII(live_metaphors_first_chapter, dormant_metaphors_first_chapter, dead_metaphors_first_chapter, metaphors_general_first_chapter)
MCI_first = MCI(everyday_metaphors_first_chapter, nature_metaphors_first_chapter, technic_metaphors_first_chapter, violence_metaphors_first_chapter, sport_metaphors_first_chapter, medical_metaphors_first_chapter, metaphors_general_first_chapter)
MPI_first = MPI(MFI_first, MII_first, MCI_first)
data_first = [RMF_first, IMC_first, MFI_first, MII_first, MCI_first, MPI_first]

df_first = pd.DataFrame(data=data_first, index=["RMF", "IMC", "MFI", "MFI", "MCI", "MPI"], dtype=float)
df_first
```

Descriptive Statistics. Second Chapter

```python
RMF_second = RMF(metaphors_general_second_chapter, words_second_chapter)
IMC_second = IMC(live_metaphors_second_chapter, dormant_metaphors_second_chapter, dead_metaphors_second_chapter, metaphors_general_second_chapter)
MFI_second = MFI(metaphors_general_second_chapter, words_second_chapter)
MII_second = MII(live_metaphors_second_chapter, dormant_metaphors_second_chapter, dead_metaphors_second_chapter, metaphors_general_second_chapter)
MCI_second = MCI(everyday_metaphors_second_chapter, nature_metaphors_second_chapter, technic_metaphors_second_chapter, violence_metaphors_second_chapter, sport_metaphors_second_chapter, medical_metaphors_second_chapter, metaphors_general_second_chapter)
MPI_second = MPI(MFI_second, MII_second, MCI_second)

data_second = [RMF_second, IMC_second, MFI_second, MII_second, MCI_second, MPI_second]

df_second = pd.DataFrame(data=data_second, index=["RMF", "IMC", "MFI", "MFI", "MCI", "MPI"], dtype=float)
df_second
```

Graphics

```python
import matplotlib.pyplot as plt
import numpy as np
array = np.array([0, 2, 4])
width = 0.8
plt.bar(array, [IMC_first, MII_first, MCI_first], label="First Chapter")
plt.bar(array+width, [IMC_second, MII_second, MCI_second], label="Second Chapter")
plt.xlabel("Quantitative Metrics") 
plt.ylabel("Value") 
plt.title("Quantitative Approach to Conceptual Metaphors (based on Dostoevsky's 'White Nights')") 
plt.xticks(array+width/2, ["IMC", "MII", "MCI"])
plt.legend() 
plt.show() 
```
